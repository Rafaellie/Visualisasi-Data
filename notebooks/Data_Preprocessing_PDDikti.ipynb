```python
# 1. Import Library

import pandas as pd
```

```python
# 2. Muat file Excel dan dapatkan semua nama lembar
file_path = '/content/sample_data/Pddikti 20 ptn.xlsx'
excel_data = pd.ExcelFile(file_path)
sheet_names = excel_data.sheet_names

# Baca semua lembar ke dalam DataFrames dan gabungkan semuanya
df_list = []
for sheet in sheet_names:
    df = pd.read_excel(file_path, sheet_name=sheet)
    df_list.append(df)

combined_df = pd.concat(df_list, ignore_index=True)
```

```python
# 3. Menampilkan informasi umum tentang DataFrame
print("Informasi DataFrame:")
combined_df.info()
```

```python
# 4. Data Preprocessing

# Menampilkan statistik deskriptif
print("\nStatistik Deskriptif:")
print(combined_df.describe(include='all'))

# Memeriksa nilai yang hilang (missing values)
print("\nJumlah Nilai yang Hilang per Kolom:")
print(combined_df.isnull().sum())

# Memeriksa nilai duplikat
print("\nJumlah Baris Duplikat:")
print(combined_df.duplicated().sum())

# Menampilkan preview data setelah preprocessing
print("\nPreview DataFrame setelah Preprocessing:")
print(combined_df.head())
```

```python
# 5. Simpan DataFrame gabungan ke file Excel baru
output_path = '/content/sample_data/Pddikti_Combined.xlsx'
combined_df.to_excel(output_path, index=False)
```
