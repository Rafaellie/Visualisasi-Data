```python
# =============================================
# Dataset: azizainunnajib/jobs-crawling (Kaggle)
# =============================================

# 1. Install & Import Library
!pip install kagglehub --quiet
!apt-get install unzip -y -qq

import os
import zipfile
import pandas as pd
import numpy as np
```

```python
# 2. Download Dataset via kagglehub
import kagglehub

path_to_zip = kagglehub.dataset_download("azizainunnajib/jobs-crawling")
print("Path to dataset ZIP:", path_to_zip)
```

```python
# 3. Ekstrak ZIP (jika file ZIP)
# Cek ekstensi; jika .zip, extract
if path_to_zip.lower().endswith(".zip"):
    extracted_dir = "jobs_crawling_dataset"
    if not os.path.exists(extracted_dir):
        os.makedirs(extracted_dir)
    print(f"Extracting zip file to: {extracted_dir}")
    with zipfile.ZipFile(path_to_zip, "r") as z:
        z.extractall(extracted_dir)
    data_dir = extracted_dir
    print("Extraction complete.")
else:
    # Jika bukan zip, asumsikan folder langsung
    data_dir = path_to_zip
    print("Downloaded path is not a zip file. Assuming it's the data directory.")
```

```python
# 4. Lihat isi folder untuk menemukan file data
print("\n=== File di dalam dataset directory ===")
for root, dirs, files in os.walk(data_dir):
    for fname in files:
        print(os.path.join(root, fname))
```

```python
# 5. Load Data ke Pandas (Gabungkan Semua File CSV)
all_files = []
# List semua file di data_dir
for root, dirs, files in os.walk(data_dir):
    for fname in files:
        if fname.endswith(".csv"):
            all_files.append(os.path.join(root, fname))

if not all_files:
    print(f"Error: Tidak ada file CSV yang ditemukan di {data_dir}")
else:
    print(f"Ditemukan {len(all_files)} file CSV. Menggabungkan...")

    # List untuk menyimpan DataFrame dari setiap file
    list_df = []

    # Baca setiap file CSV dan tambahkan ke list_df
    for csv_file in all_files:
        try:
            df_temp = pd.read_csv(csv_file)
            list_df.append(df_temp)
            print(f"Berhasil membaca: {csv_file}")
        except Exception as e:
            print(f"Gagal membaca {csv_file}: {e}")

    # Gabungkan semua DataFrame di list_df
    if list_df:
        df = pd.concat(list_df, ignore_index=True)
        print("\nSemua file CSV berhasil digabungkan.")
        print(f"Ukuran DataFrame gabungan: {df.shape}")
        # Tampilkan beberapa baris pertama dari DataFrame gabungan
        print("\nHead of combined DataFrame:")
        display(df.head()) # Menggunakan display untuk tampilan yang lebih baik di notebook
    else:
        print("\nTidak ada DataFrame yang berhasil dibaca untuk digabungkan.")
```

```python
# 6. Tampilkan Info & Preview Data
print("\n=== 5 Baris Pertama ===")
print(df.head(), "\n")

print("=== Info DataFrame ===")
print(df.info(), "\n")

print("=== Statistik Deskriptif (Numerik) ===")
print(df.describe(include="all").transpose(), "\n")
```

```python
# 7. Data Cleaning & Preprocessing

# 7.1. Cek Missing Values
print("=== Jumlah Missing Value per Kolom ===")
print(df.isna().sum(), "\n")

# 7.2. Drop Kolom yang Semua nilainya NaN atau > 80% NaN
threshold_col = len(df) * 0.8
cols_to_drop = df.columns[df.isna().sum() > threshold_col].tolist()
if cols_to_drop:
    print(f"Dropping columns karena >80% missing: {cols_to_drop}")
    df = df.drop(columns=cols_to_drop)

# 7.3. Drop Baris dengan Missing pada Kolom Penting (misal 'job_title', 'company', 'location')
important_cols = ['job_title', 'company', 'location']  # Sesuaikan dengan kolom actual
for col in important_cols:
    if col in df.columns and df[col].isna().sum() > 0:
        df = df.dropna(subset=[col])
        print(f"Dropped baris yang missing pada kolom penting: {col}")

# 7.4. Hapus Duplikasi
n_before = df.shape[0]
df = df.drop_duplicates()
n_after = df.shape[0]
print(f"\nDropped {n_before - n_after} baris duplikat\n")

# 7.5. Perbaikan Tipe Data
# Identifikasi kolom numerik vs kategorikal
numeric_cols = df.select_dtypes(include=['int64','float64']).columns.tolist()
categorical_cols = df.select_dtypes(include=['object','category','bool']).columns.tolist()

print("Kolom Numerik:", numeric_cols)
print("Kolom Kategorikal:", categorical_cols, "\n")

# Jika ada kolom numerik yang ter-read sebagai object, cek dan convert:
for col in df.columns:
    if df[col].dtype == 'object':
        # Cek apakah nilai bisa di-convert menjadi angka (misal salary)
        sample = df[col].dropna().astype(str).str.replace(',', '').str.replace('$','')
        if sample.str.match(r'^[0-9]+(\.[0-9]+)?$').all():
            df[col] = pd.to_numeric(df[col].astype(str).str.replace(',','').str.replace('$',''), errors='coerce')
            print(f"Converted column ke numeric: {col}")

# Update list kolom
numeric_cols = df.select_dtypes(include=['int64','float64','Int64']).columns.tolist()
categorical_cols = df.select_dtypes(include=['object','category','bool']).columns.tolist()

# 7.6. Imputasi Sederhana (Opsional)
# Numeric → median, Categorical → modus
for col in numeric_cols:
    if df[col].isna().sum() > 0:
        median_val = df[col].median()
        df[col] = df[col].fillna(median_val)
        print(f"Imputed missing numeric '{col}' dengan median ({median_val})")

for col in categorical_cols:
    if df[col].isna().sum() > 0:
        mode_val = df[col].mode().iloc[0]
        df[col] = df[col].fillna(mode_val)
        print(f"Imputed missing categorical '{col}' dengan mode ({mode_val})")

# 7.7. Reset Index
df = df.reset_index(drop=True)
```

```python
# 8. Simpan Data Bersih
output_clean = "jobs_crawling_cleaned.csv"
df.to_csv(output_clean, index=False)
print(f"\nData yang sudah diproses tersimpan sebagai → {output_clean}")
```
